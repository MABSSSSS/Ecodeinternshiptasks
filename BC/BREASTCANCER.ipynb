```python
# Grouping by 'label' and handling numeric and non-numeric columns separately

# 1. Grouping by 'label'
grouped = data_frame.groupby('label')

# 2. Applying different aggregations for numeric and non-numeric columns
# For numeric columns -> apply mean
# For non-numeric columns -> apply first (or any other method suitable for non-numeric data)
result = grouped.agg(
    {col: 'mean' if pd.api.types.is_numeric_dtype(data_frame[col]) else 'first' for col in data_frame.columns}
)

# Display the result
print(result)
```

```python
numeric_cols = data_frame.select_dtypes(include=['number'])

# Grouping by 'label' and calculating the mean for numeric columns
data = data_frame.groupby('label')[numeric_cols.columns].mean()
```

Seperating the features and target

```python
x = data_frame.drop(columns = 'label', axis =1)
y = data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training data and testing data

```python
x_train, x_test, y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
print("Missing values in training set:")
print(x_train.isnull().sum())
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

Model Training

Logistics Regression


```python
model = LogisticRegression()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Select numeric columns only
x_train_numeric = x_train.select_dtypes(include=[np.number])
x_test_numeric = x_test.select_dtypes(include=[np.number])

# Impute missing values with the mean
imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train_numeric)
x_test_imputed = imputer.transform(x_test_numeric)

# Scale the imputed data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_imputed)
x_test_scaled = scaler.transform(x_test_imputed)

# Train the Logistic Regression model
model = LogisticRegression()
model.fit(x_train_scaled, y_train)
```

Model evaluation

Accuracy score

```python
# Make predictions and calculate accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
print('Accuracy on training data =', training_data_accuracy)
```

```python
print('Accuracy on training data =', training_data_accuracy)
```

```python
#accuracy on training data 
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
```

```python
print('Accuracy on test data =', test_data_accuracy)
```
